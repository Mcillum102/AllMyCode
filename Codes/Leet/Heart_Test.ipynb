```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import time

def heart_shape(scale=1.0):
    t = np.linspace(0, 2 * np.pi, 100)
    x = 16 * np.sin(t)**3 * scale
    y = (13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)) * scale
    return np.append(x, x[0]), np.append(y, y[0])  # Ensure the shape is closed

def plot_heart():
    plt.ion()
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_aspect("equal")
    ax.axis("off")
    
    # Load Noto Sans Simplified Chinese font
    font_path = "NotoSansSC-Regular.otf"  # Replace with the actual path of the font
    custom_font = fm.FontProperties(fname=font_path)
    ax.set_title("", fontsize=14, fontweight='bold', fontproperties=custom_font)
    
    line, = ax.plot([], [], 'r')
    fill = ax.fill([], [], 'r', alpha=0.5)[0]
    ax.set_xlim(-20, 20)
    ax.set_ylim(-20, 15)
    
    for _ in range(50):  # Increase the number of cycles
        for scale in np.linspace(0.9, 1.1, 10).tolist() + np.linspace(1.1, 0.9, 10).tolist():
            x, y = heart_shape(scale)
            line.set_data(x, y)
            fill.set_xy(np.column_stack((x, y)))
            plt.pause(0.05)
    
    plt.ioff()
    plt.show()

if __name__ == "__main__":
    plot_heart()
```
